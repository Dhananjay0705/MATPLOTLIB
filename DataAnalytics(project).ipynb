```python
import pandas as pd
import os
from sqlalchemy import create_engine
```

```python
engine = create_engine("sqlite:///inventory.db")
```

```python
for file in os.listdir('/content/drive/MyDrive/GOOGLE COLAB FOLDER/data') :
  if '.csv' in file :
    df = pd.read_csv('/content/drive/MyDrive/GOOGLE COLAB FOLDER/data/'+file)
    print(df.shape)
    ingest_db(df,file[:-4],engine)

```

```python
def ingest_db(df,table_name,engine) :
  df.to_sql(table_name,con = engine, if_exists= 'replace',index = False)
```
